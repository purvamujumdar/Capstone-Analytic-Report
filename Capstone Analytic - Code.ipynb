```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.axes as ax
% matplotlib inline
```

```python
df = pd.read_csv('/Users/purva/Desktop/Data Science Boot Camp/AppleStore.csv')
```

```python
df.groupby('prime_genre')['prime_genre'].count()
```

```python
plt.plot(df.groupby('prime_genre')['prime_genre'].count(), color = 'blue')
plt.setp(plt.xticks()[1], rotation=30, ha='right')
figsize=(100,5)
plt.xlabel('App Genre', fontsize=14)
plt.ylabel('Number of Apps',fontsize = 14)
plt.title('Total number of apps on Apple Store per Genre',loc='Center', fontsize =18, color = 'Red')
plt.tight_layout()
plt.show()
```

```python
df.groupby('prime_genre')['price'].mean()
```

```python
df1 = df[df.prime_genre == 'Games']

## Calculating percentage of game apps of total apps

print(df1['id'].count()/df['id'].count() * 100)
```

```python
plt.hist(df1['user_rating'])
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Count of Rating',size=14)
plt.title('Distribution of rating',fontsize=18,color='red')
plt.show()
## Logarithmic distribution
```

```python

## Plotting relation between user rating and number of languages
plt.scatter(df['user_rating'], df['lang.num'])
plt.xlabel('User rating', fontsize = 14)
plt.ylabel('Available Number of languages', fontsize = 14)
plt.title('Variation of user rating against number of languages', fontsize = 18, color = 'red')
plt.show()

## Trend - User rating increases with increase in Number of languages the app is available as it increases the audience covered
```

```python
## Plotting avg price per genre for paid apps
df_free = df[(df.price == 0)]
df_nfree = df[(df.price > 10)]
df_nfree['price'].count()
plt.plot(df_nfree.groupby('prime_genre')['price'].mean())
plt.setp(plt.xticks()[1], rotation=30, ha='right')
plt.rcParams['figure.figsize']=(10,5)
plt.xlabel('App Genre',fontsize=14)
plt.ylabel('Average price for App Genre (USD)',fontsize=14)
plt.title('Average app price per Genre', fontsize=18, color = 'Red')
plt.tight_layout()
plt.show()
print(df_nfree['id'].count()/ df['id'].count())
## Education apps have highest mean price and games being app with lowest mean price
```

```python
r1 = df_nfree.sort_values(by=['price'], ascending=False).head(10)
```

```python
## Plotting top 10 most expensive apps
plt.barh(r1['track_name'], r1['price'])
plt.setp(plt.xticks()[1], rotation=30, ha='right')
plt.xlabel('Price (USD)',fontsize = 14)
plt.ylabel('Name of Apps',fontsize=14)
plt.title('Top 10 most expensive Apps', fontsize = 18, color = 'Red')
plt.tight_layout()
plt.show()
```
