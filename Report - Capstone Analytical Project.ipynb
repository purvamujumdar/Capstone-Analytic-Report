### __3. What are the trends seen in user ratings?__

After understanding the trends in the prices of mobile applications, it would be interesting to see the trends in user rating. 

The following graph shows the ditribution of user rating for all the mobile applications available on the Apple store.

From the above graph, it can be seen that there are few apps, which have a user rating of 0. So, we can say that these are the apps which might be either recently introduced and are new to the Apple store as compared to other mobile applications or do not have any ratings inspite of being old. However, it is wrong to conclude that these these might be new or do not have any rating as we do not have sufficient data about number of downloads and number of years the applications are available in the Apple store.

Apart from these mobile applications, we can see that user ratings for all other mobile applications is somewhat exponentially distributed. Although, this is not a perfect logarithmic distribution and is skewed to certain extent. However, it explains human tendency of rating an app if they like it extremely.

The user rating for any particular mobile application depends on depends on multiple factors. The following scatter plot helps us to identify the relation between the user rating and the number of language available in.

Again, assuming that the applications with user rating 0 are either new or are not rated, we can see that the user rating increases with increase in number of languages. Its most likely that the audience covered by that particular application increases with increase in the number of languages it is available in.

There is no single factor on which user rating depends. However, the above analysis can be one of the factor on which the user ratings depend.
